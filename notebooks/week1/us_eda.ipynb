```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read the cleaned data
url = https://raw.githubusercontent.com/Vincent2308/Corporate-Bankruptcy/main/data/clean/US/american_cleaned_with_names.csvdf_eda = pd.read_csv(url)
df_eda = df_eda.drop(columns=["bankrupt"], errors="ignore")
```

```python
print("Shape", df_eda.shape)
df_eda.head()
```

```python
df_eda.info()
```

```python
df_eda.describe()
```

```python
# Bankruptcy distribution
ax = df_eda["Bankrupt"].value_counts(normalize=True).plot(kind="bar")

plt.title("Bankruptcy Distribution")
plt.xlabel("Bankrupt (0 = Alive, 1 = Bankrupt)")
plt.ylabel("Proportion")

for container in ax.containers:
    ax.bar_label(container, fmt="%.2f")

plt.show()
```

```python
# Correlation matrix
numeric_df = df_eda.select_dtypes(include=['number'])

f = plt.figure(figsize=(10, 10))
plt.matshow(numeric_df.corr(), fignum=f.number)

plt.xticks(
    range(numeric_df.shape[1]),
    numeric_df.columns,
    fontsize=5,
    rotation=90
)

plt.yticks(
    range(numeric_df.shape[1]),
    numeric_df.columns,
    fontsize=5
)

cb = plt.colorbar()
cb.ax.tick_params(labelsize=8)

plt.title('Correlation Matrix', fontsize=14)
plt.show()
```

```python
"""
Sort the correlation in descending order.
Although correlation does not imply causation, it still can be a good indicator.
"""
# Select only numeric columns
numeric_df = df_eda.select_dtypes(include=['number'])

# Sort correlations with the target variable
corr_rank = numeric_df.corr()['Bankrupt'].drop('Bankrupt').abs().sort_values(ascending=False)
corr_rank
```

```python
top_features = corr_rank.head(5).index

for fea in top_features:
    sns.boxplot(x="Bankrupt", y=fea, data=df_eda)
    plt.title(f"{fea} vs Bankruptcy")
    plt.show()
```

```python
# Leverage boxplots
leverage_vars = ['long_term_debt', 'total_liabilities', 'current_liabilities', 'leverage']

for fea in leverage_vars:
    sns.boxplot(x="Bankrupt", y=fea, data=df_eda)
    plt.title(f"{fea} vs Bankruptcy")
    plt.show()
```

```python
# Profitability boxplots
profit_vars = ['net_income', 'ebitda', 'gross_profit', 'ebit_margin', 'roa']

for fea in profit_vars:
    sns.boxplot(x="Bankrupt", y=fea, data=df_eda)
    plt.title(f"{fea} vs Bankruptcy")
    plt.show()
```

```python
# Liquidity boxplots

liquidity_vars = ['current_assets', 'inventory', 'total_receivables',
                  'current_ratio', 'quick_ratio']

for fea in liquidity_vars:
    sns.boxplot(x="Bankrupt", y=fea, data=df_eda)
    plt.title(f"{fea} vs Bankruptcy")
    plt.show()
```
