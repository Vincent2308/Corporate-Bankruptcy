```python
import pandas as pd
import numpy as np
url ="https://raw.githubusercontent.com/Vincent2308/Corporate-Bankruptcy/main/data/raw/U.S/american_bankruptcy_dataset.csv"
df = pd.read_csv(url)

rename_map = {
    'X1': 'current_assets',
    'X2': 'cost_of_goods_sold',
    'X3': 'depreciation_amortization',
    'X4': 'ebitda',
    'X5': 'inventory',
    'X6': 'net_income',
    'X7': 'total_receivables',
    'X8': 'market_value',
    'X9': 'net_sales',
    'X10': 'total_assets',
    'X11': 'long_term_debt',
    'X12': 'ebit',
    'X13': 'gross_profit',
    'X14': 'current_liabilities',
    'X15': 'retained_earnings',
    'X16': 'total_revenue',
    'X17': 'total_liabilities',
    'X18': 'operating_expenses'
}

df = df.rename(columns=rename_map)

df.to_csv('american_bankruptcy_dataset_renamed.csv', index=False)

df.head()
```

```python
df['bankrupt'] = df['status_label'].map({'alive': 0, 'failed': 1})
```

```python
numeric_df = df.select_dtypes(include=['number']).copy()
```

```python
year_cols = numeric_df.columns.drop('bankrupt')

numeric_df[year_cols] = numeric_df.groupby(df['fyear'])[year_cols].transform(
    lambda x: (x - x.mean()) / x.std()
)
```

```python
corr_with_target = numeric_df.corr()['bankrupt'].sort_values(ascending=False)
print(corr_with_target)
```

```python
df = df.drop(columns=['status_label'])
```

```python
df['bankrupt'].value_counts().plot(kind='bar')
```

```python
import numpy as np
df['total_assets'] = df['total_assets'].replace(0, np.nan)
df['current_liabilities'] = df['current_liabilities'].replace(0, np.nan)
df['net_sales'] = df['net_sales'].replace(0, np.nan)
```

```python
# Compute important financial ratios
df['leverage'] = df['total_liabilities'] / df['total_assets']
df['current_ratio'] = df['current_assets'] / df['current_liabilities']
df['quick_ratio'] = (df['current_assets'] - df['inventory']) / df['current_liabilities']
df['roa'] = df['net_income'] / df['total_assets']
df['ebit_margin'] = df['ebit'] / df['net_sales']
df['gross_margin'] = df['gross_profit'] / df['net_sales']
df['retained_to_assets'] = df['retained_earnings'] / df['total_assets']
```

```python
df.to_csv('american_bankruptcy_dataset_renamed_ratios.csv', index=False)
```

```python
df.head()
```

```python
ratio_cols = ['leverage','current_ratio','roa','ebit_margin']

for col in ratio_cols:
    low, high = df[col].quantile([0.01, 0.99])
    df[col] = df[col].clip(lower=low, upper=high)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(2, 2, figsize=(14,10))

for ax, col in zip(axes.flatten(), ratio_cols):
    sns.histplot(df[col].dropna(), bins=50, kde=True, ax=ax)
    ax.set_title(col)
    ax.set_xlabel('')
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
